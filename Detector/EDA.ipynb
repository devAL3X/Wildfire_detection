# EDA (exploratory data analysis)

```python
import pandas as pd
from PIL import Image, ImageDraw
import cv2
import matplotlib.pyplot as plt
import matplotlib.patches as patches
```

```python
df = pd.read_csv("data/train/_annotations.csv", header=None, names=["img", "xmin", "ymin",        
                                                                    "xmax", "ymax", "alpha"])
```

```python
df.head()
```

```python
df["alpha"].unique()
```

"alpha" column contains only one unque value, so we can drop it

```python
df.drop([5], axis=1, inplace=True)
```

```python
def draw_random():
    row = df.sample(random_state=42).values[0]
    img = Image.open("data/train/" + row[0], mode='r')
#     img = img.convert('RGB') # to grayscale
    draw = ImageDraw.Draw(img)
    coords = [row[1], row[2], row[3], row[4]]

    fig, ax = plt.subplots(figsize=(16,9))
    ax.imshow(img)
    
    rect = patches.Rectangle(coords[:2], coords[2] - coords[0], coords[3] - coords[1], linewidth=1, edgecolor='r', facecolor='none')
    ax.add_patch(rect)
    plt.show()
```

```python
draw_random()
```
